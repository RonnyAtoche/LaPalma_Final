```python
import pandas as pd
dfns = pd.read_excel('Ronny_Presentation_NONSeyfert_Table.xlsx')
dfs = pd.read_excel('Ronny_Presentation_Seyfert_Table.xlsx')
```

```python
dfns.columns
```

```python
import matplotlib.pyplot as plt

y1 = dfns['SFR (Solar M/yr)']
y2 = dfs['SFR (Solar M/yr)']
x1 = dfns['Redshift']
x2 = dfs['Redshift']
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x1, y1, s=10, c='b', marker="s", label='NON-Seyfert')
ax1.scatter(x2,y2, s=10, c='r', marker="o", label='Seyfert')
plt.legend(loc='upper left')
plt.title('SFR')
plt.xlabel('Redhsift')
plt.ylabel(r'$ SFR(M\odot yr^{-1}) $')
plt.show()
```

```python
dfnsSFR = dfns['SFR (Solar M/yr)']
dfsSFR = dfs['SFR (Solar M/yr)']
```

```python
fig, ax = plt.subplots()
ax.boxplot([dfsSFR, dfnsSFR])
plt.title('SFR')
plt.xlabel('Seyfert / NON-seyfert')
plt.ylabel(r'$ SFR(M\odot yr^{-1}) $')
plt.show()
```
